```python

import os.path
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

import urllib.request
import time
import ssl
from selenium.common import exceptions

ssl._create_default_https_context = ssl._create_unverified_context

URL = r"d:\laiba Shadi\untitled folder 5\error\._QUE05429.JPG"

IMG_COUNT = 5


def set_chrome_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("headless")  
    options.add_argument("lang=ko_KR") 
    options.add_experimental_option(
        "excludeSwitches", ["enable-logging"]
    ) 
    options.add_argument("start-maximized")
    options.add_argument("disable-infobars")
    options.add_argument("--disable-extensions")

    driver = webdriver.Chrome(
        service=Service(ChromeDriverManager().install()), options=options
    )
    return driver


def get_events():
    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")
   
    idx = 1
    while True:
        html_events = driver.find_elements(By.XPATH, '// *[ @ id = "bbsList"] / li')

        print("===================")
        print(f"{idx} count : " + str(len(html_events)))
     
        idx += 1
        if len(html_events) >= 1000:
            break
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load page
        time.sleep(3)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    return html_events


if __name__ == "__main__":

    start = time.time()  

    data = []

    driver = set_chrome_driver()
    driver.get(URL)

    html_events = get_events()

    # save img files to disk
    for event in html_events:
        try:

            print(event.text)
            split_text = event.text.split("\n")
            # title_period
            save_path = "./images/" + split_text[0]
            href = event.find_element(By.CSS_SELECTOR, 'a')
            href.get_attribute('href')
            href.click()

            images = driver.find_elements(By.XPATH, f' // *[ @ id = "etcimgSec"] / div[1] / div[2] / div / li /img')
            for i, img in enumerate(images):
                if i == IMG_COUNT:
                    break
                url = img.get_attribute('src')
                if not os.path.isdir(save_path):
                    os.makedirs(save_path)
                urllib.request.urlretrieve(url, save_path + f"/{i}.jpg")
            print("saved images to " + save_path)
            print("===================")
        except exceptions.StaleElementReferenceException as e:
            print(f"An error occurred: {e}")
        except exceptions.NoSuchElementException:
            print("Element not found")
            break
        except Exception as e:
            print(f"An error occurred: {e}")
            break

    end = time.time()  
    print(f"{end - start:.5f} sec")
```

```python
import pygame

pygame.init()
screen = pygame.display.set_mode((1000, 480)) 

# Image paths
img_paths = [
    "image/images (1).jpg",
    "image/images (2).jpg",
    "image/images (3).jpg",
    "image/images (4).jpg",
    "image/download (1).jpg",
    "image/download (2).jpg",
    "image/download.jpg",
]

# Load images
images = [pygame.image.load(path).convert() for path in img_paths]

num_rows = len(images)
num_cols = 1  
for row in range(num_rows):
    for col in range(num_cols):
        index = row * num_cols + col
        if index < len(images):
            x_position = col * images[index].get_width()
            y_position = row * images[index].get_height()
            screen.blit(images[index], (x_position, y_position))

pygame.display.flip()
pygame.time.delay(3000) 

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

pygame.quit()
```

```python
import os
import pygame

pygame.init()
screen = pygame.display.set_mode((800, 600))  
folder_path = "image"

# List all files in the folder
all_files = os.listdir(folder_path)

# Filter for image files (you might want to adjust this based on your file types)
image_files = [file for file in all_files if file.lower().endswith((".jpg", ".png", ".jpeg"))]

# Take the first three image files
selected_images = image_files[:3]

# Load images
images = [pygame.image.load(os.path.join(folder_path, image)).convert() for image in selected_images]

# Calculate the maximum width and height among the images
max_width = max(img.get_width() for img in images)
max_height = max(img.get_height() for img in images)

# Calculate the number of columns and rows based on the screen size
num_columns = screen.get_width() // max_width
num_rows = screen.get_height() // max_height

# Calculate the gap between images
horizontal_gap = (screen.get_width() - (num_columns * max_width)) // (num_columns - 1)
vertical_gap = (screen.get_height() - (num_rows * max_height)) // (num_rows - 1)

# DisplayX images in a grid
for row in range(num_rows):
    for col in range(num_columns):
        index = row * num_columns + col
        if index < len(images):
            x_position = col * (max_width + horizontal_gap)
            y_position = row * (max_height + vertical_gap)
            screen.blit(images[index], (x_position, y_position))

pygame.display.flip()

pygame.time.delay(3000)

pygame.quit()
```
